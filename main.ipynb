```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("energy.csv")
```

```python
#removes the null values
data.dropna(inplace = True)
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['EnergyConsumption'], axis=1 )
y = data['EnergyConsumption']
```

```python
x
```

```python
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size= 0.25 , random_state = 42)
```

```python
train_data = x_train.join(y_train) 
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.corr
```

```python
# Convert categorical columns to numerical values using encoding
encoded_data = pd.get_dummies(train_data.drop(columns=['Timestamp']), drop_first=True)
plt.figure(figsize=(15, 8))
sns.heatmap(encoded_data.corr(), annot=True, cmap="YlGnBu")
plt.show()
```

```python
# implemwnt incase skewness exists 
train_data['Temperature'] = np.log(train_data['Temperature'])
train_data['Humidity'] = np.log(train_data['Humidity'])
train_data['EnergyConsumption'] = np.log(train_data['EnergyConsumption'])
```
